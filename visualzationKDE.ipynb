```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
ds = randn(290)
```

```python
sns.kdeplot(ds)
```

```python
# We can adjust the bandwidth of the sns kde to make the kde plot more or less sensitive to high frequency

# Rugplot
sns.rugplot(ds, color='red')

#plot with various bandwiths

for bw in np.arange(0.5,2,0.25):
    sns.kdeplot(ds,bw=bw,lw=1.8, label = bw)
```

```python
# now choose different kernels

kernel_options = ["biw", "cos", "epa", "gau", "tri", "triw"]

# for further info the link below
url = 'http://en.wikipedia.org/wiki/Kernel_(statistics)'

# Use label to set legend
for kern in kernel_options:
    sns.kdeplot(ds,kernel=kern,label=kern)
```

```python
# we can also shade if desired

for kern in kernel_options:
    sns.kdeplot(ds, kernel=kern, label= kern, shade =True, alpha=0.5)
```

```python
sns.kdeplot(ds, vertical=True)
```

Multivariate Density Estimation using KDEPLOT

also using KDE for multidimensional data

```python
#creating new dataset
#mean center of data
mean=[0,0]

#diagonal ds using numpy

cov = [[1,0],[0,100]]

#create ds using numpy
ds2 = np.random.multivariate_normal(mean,cov,1000)

#creating df with panda

df = pd.DataFrame(ds2, columns=['X','Y'])

#plot df

sns.kdeplot(df)
```

```python
# We could have also passed two vectors seperately, and shade
sns.kdeplot(df.X,df.Y,shade=True)
```

```python
sns.kdeplot(df,bw=1)
```

```python
sns.kdeplot(df,bw='silverman')
```

```python
#combine ds with jointplot

sns.jointplot('X','Y',df,kind='kde')
```
