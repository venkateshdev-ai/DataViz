```python
# libraries required 

import numpy as np
from pandas import Series, DataFrame
import pandas as pd
from numpy.random import randn
#import stats from numpy

from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook if u r using Jupyter Notebook
%matplotlib inline
```

```python
#Create a random normal-dist dataset
ds1 = randn(100)

#Plot a histogram of the dataset, note bins=10 by default
plt.hist(ds1)
```

```python
ds2 = randn(80)


plt.hist(ds2, color='indianred')
```

```python
# We can use "density" to plot on same plot

# Set densit(instead of normed)=True for the plots to be normalized in order to comapre data sets with different number of observations
# Set alpha=0.5 for transperancy

plt.hist(ds1,density=True,color='indianred',alpha=0.5,bins=20)
plt.hist(ds2,density=True,alpha=0.5,bins=20)
```

```python
#now again making randn dataset and using join plot for visual to make scatter plot

data1 = randn(1000)
data2 = randn(1000)

sns.jointplot(data1,data2)
```

```python
# Can also use hex bins for a more concise picture
sns.jointplot(data1, data2, kind='hex')
```
